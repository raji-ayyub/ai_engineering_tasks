### Question 1d. :

* `np.linalg.det()` → determinant
* `np.linalg.inv()` → inverse (only works if determinant ≠ 0)

```python
import numpy as np

matrix_A = np.array([[2, -1],
                     [0, 3]])

matrix_B = np.array([[1, -2],
                     [4, 5]])

# Determinants
det_A = np.linalg.det(matrix_A)
det_B = np.linalg.det(matrix_B)

print("Determinant of A:", det_A)
print("Determinant of B:", det_B)

```


###  Results:

* **Determinant of A and B:**

$$
\text{det} = (2)(3) - (0)(-1) = 6
$$

$$
\text{det} = (5)(1) - (4)(-2) = 13
$$

both ≠ 0 → Hence are invertible.

### Transpose of A and B:

* **Inverse formula for a 2×2 matrix**:

$$
\text{If } M = \begin{bmatrix} a & b \\ c & d \end{bmatrix}, \quad 
M^{-1} = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$


$$
A^{-1} = \frac{1}{6} \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}
= \begin{bmatrix} 1/3 & -1/6 \\ 0 & 1/2 \end{bmatrix}
$$


$$
B^{-1} = \frac{1}{13} \begin{bmatrix} 1 & 4 \\ -2 & 5 \end{bmatrix}
= \begin{bmatrix} 1/13 & 4/13 \\ -2/13 & 5/13 \end{bmatrix}
$$
